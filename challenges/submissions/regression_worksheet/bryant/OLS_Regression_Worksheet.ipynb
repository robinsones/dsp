```python
df = pd.DataFrame({'P':P, 'SVol':SVol, 'Ing':Ing})
```

```python
from patsy import dmatrices
patsy_formula = 'SVol ~ Ing'
y, x = dmatrices(patsy_formula, data=df, return_type='dataframe')
```

```python
import statsmodels.api as sm
model = sm.OLS(y, x)
results = model.fit()
results.summary()
```
