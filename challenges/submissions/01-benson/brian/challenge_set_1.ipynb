# Challenge 1

```python
import pandas as pd
import csv
import ggplot
df = pd.read_csv('/home/brianc/Documents/2016_06_Week4_Turnstyle_Data')
df.head(n=30)
```

# Challenge 2

```python
df2 = df[['C/A','UNIT','SCP','STATION','DATE','TIME','ENTRIES']]
df2.head(n=10)
```

# Challenge 3

```python
def range(group):
    entryCount = group.max()-group.min()
    return entryCount
df3 = df2.groupby(['C/A','UNIT','SCP','STATION','DATE'])['ENTRIES'].apply(range)
df3 = pd.DataFrame(df3)
df3.head(n=20)
```

# Challenge 4

```python
ggplot(aes(x='DATE',y='ENTRIES'),data=df3)+geom_line()
```
