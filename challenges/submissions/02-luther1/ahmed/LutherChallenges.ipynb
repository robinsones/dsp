```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Challenge 1

```python
df = pd.read_csv('/Users/El-Razi/ds/metis/nyc16_ds8/challenges/challenges_data/2013_movies.csv')
df['ReleaseDate'] = pd.to_datetime(df.ReleaseDate)
```

```python
df['DomesticTotalGross'] /= 1000000
```

```python
df['Date'] = df.ReleaseDate.dt.month
```

```python
f = plt.figure()
f.set_size_inches(15,4)
ax = plt.axes()
ax = df.plot('ReleaseDate','DomesticTotalGross', legend=False, ax=ax)
ax.set_ylabel('Domestic Total Gross')
ax.set_xlabel('Release Date')
```

# Challenge 2

```python
f = plt.figure()
f.set_size_inches(15,4)
ax = plt.axes()
ax = df.plot('Runtime','DomesticTotalGross', legend=False, ax=ax, kind='scatter')
#plt.plot(df.sort('Runtime')['Runtime'], df.sort('Runtime')['DomesticTotalGross'])
ax.set_xlabel('Runtime')
ax.set_ylabel('Domestic Total Gross (Millions)')
```
