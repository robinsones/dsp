# Challenge 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('2013_movies.csv')
df['ReleaseDate'] = pd.to_datetime(df.ReleaseDate)
```

```python
df.plot('ReleaseDate', 'DomesticTotalGross',kind='line', figsize = (20,10), fontsize= 18)
```

# Challenge 2

```python
df = df.sort_values(by='Runtime')
```

```python
df.plot('Runtime', 'DomesticTotalGross',kind='bar', figsize=(20,10), fontsize= 18)
```

# Challenge 3

```python
by_rating = df.groupby('Rating')['DomesticTotalGross','Runtime'].mean()
```

```python
by_rating = by_rating.reset_index()
by_rating
```

# Challenge 4

```python
by_rating_dtg = df.groupby('Rating')
```

```python
ax = plt.axes()

by_rating_dtg.plot('ReleaseDate', 'DomesticTotalGross',kind='line',figsize = (10,10), fontsize= 18, ax=ax)
legend = ax.legend(loc=0)
rating_labels = ['G','PG','PG-13','R']
_ = [x[1].set_text(rating_labels[x[0]]) for x in enumerate(legend.get_texts())]
```
