```python
import numpy as np

int_matrix = (2, 7, 1, 5, 10)
candidates = np.arange(0, 10, .1)
sum_matrix = np.zeros(100)
```

```python
min_results = 100000000
candidate = 0.0
index = 0
```

```python
for tenth in candidates:
    temp_results = []
    for integer in int_matrix:
        temp_results.append( (integer - tenth)**2 )
        sum_matrix[index] = sum(temp_results)
    index += 1
```

```python
idx = np.argmin(sum_matrix)
print( candidates[idx], sum_matrix[idx] )
```
